## Exercise 1

```python
def positionWord(filePath, position):
  result=[]
  
  with open(filePath, "r") as readfile:  #open file
    Text=readfile.read()                 #read file,lab1
  
    lines=Text.splitlines()              #lab1, splitting the text to a list of lines using splitlines()
    
    for words in lines:
        #print (words)
        wordsp=words.split()             #splitting the text to a list of words using split()
        #print (wordsp)  
        result.append(wordsp[position].strip(""" ,.*()[]!@#$%^&*{}?'`"-"""))

  return result
```

```python
positionWord("example.txt", 3)
```

```python
def onlyUpperCase(text):
  result = ''                #define a string called result, will be used in final return
  for i in text:
    if i.isupper():          #check if it is uppercase
      result = result + i    #combine two strings to one string
  return result              #return the string
```

```python
onlyUpperCase("lorem ipsum Dolor Sit amet, conSectetur.")
```

## Exercise 3

```python
def divisors(num1, num2):
  result = []                     #it is used to store new list
  for i in range(1,num1+1):
      if i % num2 == 0:           #check whether there is a remainder or not
        result.append(i)          #add i to the list of factors
  return result                   #return the entire list
```

```python
divisors(23, 6)
```

```python
from collections import Counter            #call library//lab1
def countWords(WordList):
  Wordcounter={}                           #empty dictionary for future retrun
  Wordcounter=Counter(WordList)
  
  return dict(Wordcounter)
```

```python
countWords(["the", "seething", "sea", "ceaseth",
         "and", "thus", "the", "seething", "sea", "sufficeth", "us"])
```

## Exercise 5

```python
from collections import Counter
def organizedCountWords(wordList):
  by_wordlength={}                  #empty dictionary for result return
  Wordcounter=Counter(wordList)     #same with previous question; counting words
  
  

  for word in wordList:
    if len(word) in by_wordlength:
      by_wordlength[len(word)][word]=Wordcounter[word]
     
    else:
      by_wordlength[len(word)]={word:Wordcounter[word]}#num:{word:counter number}
  
  
  return by_wordlength
```

```python
organizedCountWords(["the", "seething", "sea", "ceaseth",
         "and", "thus", "the", "seething", "sea", "sufficeth", "us"])
```

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/MIE451-1513-2019/course-datasets/master/toronto_airbnb_listings_Aug2019.csv'
obj=pd.read_csv(url)
import numpy as np

def AvgNeighbourhoodListingPrice(neighbourhoodList, room_type, number_of_reviews, minimum_nights):
    
    '''
    Neighbour type:list[str]
    room_type type: str
    minimum review type:int
    nights type: int 
    '''
   
  
    aa=obj[obj['minimum_nights']<=minimum_nights]   #making comparation
    bb=aa[aa['number_of_reviews']>=number_of_reviews]
    
    cc=bb[bb['neighbourhood'].isin(neighbourhoodList)]#true / false, in this List?
    dd=cc[cc['room_type']==room_type]

    grouped1 = dd[["neighbourhood","price"]]
    grouped2 = grouped1.groupby(["neighbourhood"],as_index=True).mean()
    grouped3 = grouped2.sort_values(by="price",ascending=False).round(2)
    #sorting data, pass one column named'price',the data is sorted
    #the data is sorted in descending order
    
    
    return grouped3
  
```

```python
AvgNeighbourhoodListingPrice(["Edenbridge-Humber Valley","Annex", "The Beaches"],"Entire home/apt", 30, 3)

```

```python
import numpy as np
def leastSquaresFit(URLX,URLy):
        
    '''
    input str 
    URLX type: matrix[str]
    URLY type: matrix[str]
    return type: ndArray[float](3,1)
    '''
    import pandas as pd

    # change the URL to URLx and URLy.
    x=pd.read_csv(str(URLX),header=None).astype('float')
    x['3']=1#insert c0 in original data
    #print (x)
    y=pd.read_csv(str(URLy), header=None).astype('float')
    #convert or cast an array from one dtype to another by using .astype
    
    
    #form matrix
    x_np=x.values#.as_matrix(columns=None)
    y_np=y.values#.as_matrix(columns=None)
    
    w=np.matmul(np.matmul(np.linalg.inv(np.matmul(x_np.T,x_np)),x_np.T),y_np)
    #matmul returns the matrix product of two arrays, numpy.linalg.inv computes inverse of matrix
    w=w[0:3]#last one is c0, no need, only get first three rows
    return w
```

```python
leastSquaresFit('https://raw.githubusercontent.com/MIE451-1513-2019/course-datasets/master/X.csv','https://raw.githubusercontent.com/MIE451-1513-2019/course-datasets/master/Y.csv')
```
